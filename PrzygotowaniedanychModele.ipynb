## Braki danych

```python
pd.set_option('display.max_rows', None)
data[data.isnull().any(axis=1)]
```

```python
def whatPercent(data, col):
    lenRows = len(data[col])
    lenNan = len(data[data[col].isna()])
    if data[col].isna().any():
        return f'{round((lenNan / lenRows) * 100, 2)}%'
```

```python
data['heartRate'].isna().any()
```

```python
for i in data.columns:
    if data[i].isna().any():
        print(f'{i:10} -->  {whatPercent(data,i)}')
```

ile jest braków danych w każdej kolumnie na ilość wierszy. 

```python
data.education.fillna(method = 'ffill',inplace = True)
data.heartRate.fillna(method = 'ffill',inplace = True)
```

education jest kategoryczna, więc wypełniłam puste komórki wartością z poprzedniego wiersza, a heartRate dlatego, że brakuje tylko jednej wartości.

```python
data[data['currentSmoker'] == 1]['cigsPerDay'].isna().sum()
```

czyli wyszło, że braki zdarzają się tylko w sytuacji kiedy osoby palącej. Więc policzymy średnią ale biorąc tylko pod uwagę palaczy. 

```python
meanCigs = round(data[data['currentSmoker'] == 1]['cigsPerDay'].mean())
meanBMI = round(data['BMI'].mean(),2)
meanGlu = round(data['glucose'].mean(),2)

data['cigsPerDay'].fillna(value = meanCigs, inplace = True)
data['totChol'] = data['totChol'].apply(lambda x: round(np.mean(data['totChol'])) if np.isnan(x) == True else x)
data['BMI'].fillna(value = meanBMI, inplace = True)
data['glucose'].fillna(value = meanGlu, inplace = True)
```

te cztery kolumny wypełniłam średnimi

```python
notMeds = data['BPMeds'] == 0
yesHyp = data['prevalentHyp'] == 1
lowSysBP = data['sysBP'] < 120
highDiaBP = (data['diaBP'] < 90) & (data['diaBP'] > 60)
data[notMeds & yesHyp & lowSysBP & highDiaBP][['age','sysBP','diaBP']]
```

```python
lowSysBP = data['sysBP'] < 130
highDiaBP = data['diaBP'] > 90
data[notMeds & yesHyp & lowSysBP & highDiaBP][['age','sysBP','diaBP']]
```

```python
yesHyp = data['prevalentHyp'] == 1
data[yesHyp]['BPMeds'].unique()
```

```python
model = LogisticRegression()
dataPrep = data[['BPMeds','prevalentHyp','sysBP','diaBP']]
testdf = dataPrep[dataPrep['BPMeds'].isnull()==True]
traindf = dataPrep[dataPrep['BPMeds'].isnull()==False]
y = traindf['BPMeds']
traindf.drop(["BPMeds"],axis=1,inplace=True)
model.fit(traindf,y)
testdf.drop(["BPMeds"],axis=1,inplace=True)
pred = model.predict(testdf)
testdf['BPMeds'] = pred
```

```python
data['BPMeds'].fillna(testdf['BPMeds'], inplace = True)
data
```

Uzywamy regresji logistycznej do imputacji braków danych w kolumnie BPMeds. Zmiennymi objaśniającymi jest prevalentHyp, sysBP, diaBP.  

```python
categories = [{'sex': {1:'male', 0:'female'}},
{'currentSmoker' : {0:'No', 1:'Yes'}},
{'BPMeds': {0:"No", 1:"Yes"}},
{'prevalentStroke' : {0:"No", 1:"Yes"}},
{'prevalentHyp' : {0:"No", 1:"Yes"}},
{"diabetes" : {0:"No", 1:"Yes"}},
{"TenYearCHD" : {0:"No", 1:"Yes"}}]
```

```python
for i in categories:
    key_values = i.items()
    data.replace({str(key): value for key, value in key_values}, inplace = True) 
```

```python
for i in data.columns:
    if data[i].nunique() < 5:
        data[i] = data[i].astype('category')
```
